#**Random Tensors**

###Why random tensors ?
###Random tensors are important because

Start with random tensors  -> look at data -> update random numbers -> look at data -> update again -> Keep repeating this till the final epoch


Random tensors : https://docs.pytorch.org/docs/stable/generated/torch.rand.html

```python
random_scalar = torch.rand(4)
random_scalar
```

```python
random_matrix = torch.rand(2,3)
random_matrix
```

```python
random_tensor = torch.rand(1,3,4)
random_tensor
```

```python
random_tensor.ndim
```

```python
# create a random tensor for image
random_image_size_tensor = torch.rand(size=(3,224,224))
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

#**Creating a range of tensors and tensor like**

arange: https://docs.pytorch.org/docs/stable/generated/torch.arange.html

tensor_like: https://docs.pytorch.org/docs/stable/generated/torch.zeros_like.html

```python
one_to_ten = torch.arange(1,11)
one_to_ten
```

```python
one_to_thousand = torch.arange(start=0, end=1001, step=100)
one_to_thousand
```

```python
one_to_ten_like = torch.zeros_like(one_to_ten)
one_to_ten_like
```

#**Tensor Data Type**

**Note: ** This is going to be one of the 3 most common errors
1. Tesors with inccorect data type
2. Tesors with inccorect shape
3. Tesors with inccorect device

```python
float_32_tensor = torch.tensor([3.0,5.0,2.0], device= None, dtype=torch.float32, requires_grad= False )
float_32_tensor
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_32_tensor * float_16_tensor
```

#**Tensor Attributes**
1. shape
2. device
3. dtype

```python
# find out the attributes of the tensor
print(f"Datatype of tensor: {float_32_tensor.dtype}")
print(f"Shape of tensor: {float_32_tensor.shape}")
print(f"Device of tensor: {float_32_tensor.device}")
```
