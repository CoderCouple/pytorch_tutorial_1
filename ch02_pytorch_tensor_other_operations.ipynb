# Tensor transformation
1. Reshaping
2. Stacking
3. Squeezing
4. Unsqueezing



```python
# create a tensor
import torch
x = torch.arange(1., 11.)
x, x.shape
```

```python
# add an extra dimension
x_reshaped = x.reshape(1, 10)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(2, 5)
x_reshaped, x_reshaped.shape
```

```python
# change the view (view (Z) shares the memory with the tensor (X) )
# changeing Z will chaneg X and vice versa.
z = x.view(1, 10)
z, z.shape
```

```python
# stack tesnsors on eachother
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```
